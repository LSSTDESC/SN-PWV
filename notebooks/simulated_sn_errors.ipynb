# Visualizing SN Lightcurves


```python
import os
import sys

import sncosmo
from astropy.table import Table
from matplotlib import pyplot as plt

sys.path.insert(0, '../')
from snat_sim import filters, plasticc, constants, lc_simulation

os.environ['CADENCE_SIMS'] = '/mnt/md0/sn-sims/'
filters.register_lsst_filters(force=True)
```

```python
cadence = 'alt_sched'
model = 11
plastic_iter = plasticc.iter_lc_for_cadence_model(cadence, model, verbose=False)
```

```python
def plot_next_lc():
    plasticc_lc = next(plastic_iter)

    formatted_lc = plasticc.format_plasticc_sncosmo(plasticc_lc)
    plasticc_fig = sncosmo.plot_lc(formatted_lc)
    plasticc_fig.suptitle('PLaSTICC Lightcurve', y=1.05, fontsize=20)
    plt.show()

    duplicated_lc = plasticc.duplicate_plasticc_sncosmo(plasticc_lc, sncosmo.Model('salt2-extended'))
    sncosmo_fig = sncosmo.plot_lc(duplicated_lc)
    sncosmo_fig.suptitle('sncosmo Lightcurve', y=1.05, fontsize=20)
    plt.show()
```

```python
plot_next_lc()
```
